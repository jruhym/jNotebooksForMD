```python
import bioinf as bi

file = 'Nap_everynsInteriorSolvent_30deg.pdb'
nonProteinResidues = ('DPPC','SOL','NA', 'CL', 'NAP', 'CMYH')
mainChain = 'A'

with open(file) as f, open('corrected.pdb', 'w') as w:
    for line in f:
        cleanline = line.strip()
        if not cleanline.startswith("ATOM"):
            w.write(f'{cleanline}\n')
            continue
        (ATOM, serial, name, resName_raw, resSeq, x, y, z, occupancy, tempFactor, element) = cleanline.split()
        resName = resName_raw.translate({ord(f'{a}'):None for a in range(10)})
        resSeq_from_resName = resName_raw[len(resName):]
        resSeq_from_resName_int = -1 if len(resSeq_from_resName) == 0 else int(resSeq_from_resName)
        resSeq = f'{max(resSeq_from_resName_int, int(resSeq))}'
        
        atom_line = bi.PDBAtomLine(serial, name, '', resName,
                                   '' if resName in nonProteinResidues else mainChain,
                                   resSeq, '', x, y, z, occupancy, tempFactor, element, '', kind=ATOM)
        if serial == '5396':
            print(atom_line)
        w.write(f'{str(atom_line)} \n')
```
